```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# GitHub URL of the CSV file
github_url = ('https://raw.githubusercontent.com/Afnan-afy/Covid-19-Data-Visualization/main/country_wise_latest.csv')

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(github_url)

# Display the DataFrame
print(df)
```

```python
import pandas as pd

# Assuming your DataFrame is named df
# Group the data by "WHO Region"
grouped = df.groupby("WHO Region")

# Calculate aggregate statistics for each group
aggregate_statistics = grouped.agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
    'Active': 'sum',
    'New cases': 'sum',
    'New deaths': 'sum',
    'New recovered': 'sum',
    'Deaths / 100 Cases': 'mean',  # Calculating mean for demonstration purposes
    'Recovered / 100 Cases': 'mean',
    'Deaths / 100 Recovered': 'mean',
    'Confirmed last week': 'sum',
    '1 week change': 'sum',
    '1 week % increase': 'mean'
})

# Display the aggregate statistics
print(aggregate_statistics)
```

```python
import pandas as pd

# Example DataFrame
# df = ...

# Descriptive statistics
descriptive_stats = df.describe()
print(descriptive_stats)

# Grouping and aggregation
grouped_data = df.groupby('WHO Region').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'
})
print(grouped_data)
```

```python
df.groupby('WHO Region')['Deaths'].sum().plot(kind='bar', title=" Death based on region")
plt.xlabel('WHO Region')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
```

```python
df.groupby('WHO Region')['Recovered'].sum().plot(kind='pie', title=" Recovered based on region",figsize =(10,6),autopct='%1.1f%%')
```

```python
df.groupby('WHO Region')['Deaths'].sum().plot(kind='pie', title=" Deaths based on region",figsize =(10,6),autopct='%1.1f%%')
```

```python
df.groupby('WHO Region')['Active'].sum().plot(kind='line', title=" Active cases based on region",figsize=(10,6))
```

Show me the Confirmed,Deaths,Recovered & Active case in scattered plot

```python
sns.pairplot(df[['Confirmed', 'Deaths', 'Recovered', 'Active']])
plt.title('Pairplot of Confirmed, Deaths, Recovered, and Active Cases')
plt.show()
```

show me the top 10 confirmed case from the entire country

```python
top_10_confirmed = df.sort_values(by='Confirmed', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x='Confirmed', y='Country/Region', data=top_10_confirmed)
plt.xlabel('Confirmed Cases')
plt.ylabel('Country/Region')
plt.title('Top 10 Countries with Most Confirmed Cases')
plt.show()
```

Show the confirmed and Recovered case 

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Confirmed', y='Recovered', data=df)
plt.xlabel('Confirmed Cases')
plt.ylabel('Recovered Cases')
plt.title('Confirmed vs. Recovered Cases')
plt.show()
```

```python
india_data = df[df['Country/Region'] == 'India'][['Confirmed' , 'Deaths' , 'Recovered' , 'Active' ]]

print(india_data)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(india_data.columns, india_data.iloc[0].values, color=['blue', 'red', 'green', 'orange'])
plt.title('COVID-19 Cases in India')
plt.xlabel('Case Type')
plt.ylabel('Number of Cases')
plt.show()
```
