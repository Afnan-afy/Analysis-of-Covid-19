```python
import pandas as pd

# Assuming your DataFrame is named df
# Group the data by "WHO Region"
grouped = df.groupby("WHO Region")

# Calculate aggregate statistics for each group
aggregate_statistics = grouped.agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
    'Active': 'sum',
    'New cases': 'sum',
    'New deaths': 'sum',
    'New recovered': 'sum',
    'Deaths / 100 Cases': 'mean',  # Calculating mean for demonstration purposes
    'Recovered / 100 Cases': 'mean',
    'Deaths / 100 Recovered': 'mean',
    'Confirmed last week': 'sum',
    '1 week change': 'sum',
    '1 week % increase': 'mean'
})

# Display the aggregate statistics
print(aggregate_statistics)
```

```python
import pandas as pd

# Example DataFrame
# df = ...

# Descriptive statistics
descriptive_stats = df.describe()
print(descriptive_stats)

# Grouping and aggregation
grouped_data = df.groupby('WHO Region').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'
})
print(grouped_data)
```

```python
df.groupby('WHO Region')['Deaths'].sum().plot(kind='bar', title=" Death based on region")
plt.xlabel('WHO Region')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
```

```python
df.groupby('WHO Region')['Recovered'].sum().plot(kind='pie', title=" Recovered based on region",figsize =(10,6),autopct='%1.1f%%')
```

```python
df.groupby('WHO Region')['Deaths'].sum().plot(kind='pie', title=" Deaths based on region",figsize =(10,6),autopct='%1.1f%%')
```

```python
df.groupby('WHO Region')['Active'].sum().plot(kind='line', title=" Active cases based on region",figsize=(10,6))
```
